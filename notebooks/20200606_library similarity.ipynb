# 20200606 Library similarity calculations

## Goals
* Given a randomly, under-sampled library, what is the closest we can expect 2/3/4/5/... etc sequences to be within it?

## Thoughts / considerations / approaches

### Enrichment above basal library bias
* instead of looking for similar sequences, look for AAs (or groups of AAs?) that are enriched above what would be expected from the NNS library bias)
* would only be meaningful if multiple sequences hit the same modality
    * could perhaps bin sequences into their modalities using this
    
#### Library bias
* in NNK and NNS libraries, the theoretical ratio between Arg/Leu/Ser and Met/Trp is 81:1

### Example libary 1:


* $1.4*10^{14}$ nucleotide variants
* $2*10^{12}$ peptide variants ($4*10^{12}$ including stop codons)

### Example library 2:


* $1.4*10^{11}$ nucleotide variants
* $5.2*10^{9}$ peptide variants ($7.2*10^{9}$ including stop codons)
